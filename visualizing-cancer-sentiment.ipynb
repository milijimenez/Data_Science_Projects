# **3. Data Analysis**
The analysis consist of:
* countplot
* wordcloud

**Question: What is the distribution of the intensity scores among the posts by the users**

```python
fig = plt.figure(figsize=(20,8),facecolor='white')
colors = ["#b2d8d8","#66b2b2","#008080","#006666"]
sns.countplot(x=data['intensity'], palette=colors)
plt.title('Distribution of the Intensity Scores Across the Data')        
plt.show()
```

**Question: What is the most commmon words used by the users in their posts?**

```python
from wordcloud import WordCloud
check_titles = ''.join(map(str,data['posts']))
wordcloud_title = WordCloud(width = 800, height = 800,
                            background_color ='white',
                            min_font_size = 10).generate(check_titles)
plt.figure(figsize=(6,6))
plt.imshow(wordcloud_title)
plt.axis('off')
plt.title('User Review Title', fontsize=20)
plt.tight_layout(pad=2)
plt.show()
```
